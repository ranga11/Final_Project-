### ARIMA

```python
import matplotlib
import matplotlib.pylab as plt
import matplotlib.finance as mpf
matplotlib.style.use('seaborn')
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 5
from plotly.graph_objs import *
from plotly.offline import init_notebook_mode, iplot, iplot_mpl
init_notebook_mode()
from tqdm import tqdm
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

```python
df2=df2.rename(index=str, columns={"index": "Date"})
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
df2 = df2.set_index('Date')
```

```python
df2['Close_diff'] = df2['AAL_Close']-df2.shift()['AAL_Close']
```

```python
df2['Close_diff_log'] = np.log1p(df2['AAL_Close'])-np.log1p(df2.shift()['AAL_Close'])
```

```python
df2 = df2[['AAL_Close', 'Close_diff', 'Close_diff_log']]
df2 = df2.dropna()
df2.head()
```

```python
train = df2['2013-12-10':'2017-06-01']
test = df2['2017-06-01':]
len(train), len(test)
```

```python
iplot([Scatter(x=train.index, y=train['Close_diff'])])
```

```python
iplot([Histogram(x=train['Close_diff'])])
```

```python
plt = sm.graphics.tsa.plot_acf(train['Close_diff'], lags=40)
plt.show()
```

```python
plt = sm.graphics.tsa.plot_pacf(train['Close_diff'], lags=40)
plt.show()
```

```python
res = sm.tsa.arma_order_select_ic(train['Close_diff'], ic='aic', trend='nc')
res
```

```python
from statsmodels.tsa.arima_model import ARIMA

arima_3_1_0 = ARIMA(train['AAL_Close'].as_matrix(), order=(3, 1, 0)).fit(dist=False)
arima_3_1_0.summary()
```

```python
# plot residual errors
residuals = pd.DataFrame(arima_3_1_0.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

```python
plt = sm.graphics.tsa.plot_acf(arima_3_1_0.resid, lags=40)
plt.show()
```

```python
plt = sm.graphics.tsa.plot_pacf(arima_3_1_0.resid, lags=40)
plt.show()
```

```python
ts = train['AAL_Close'].as_matrix()
predictions = np.empty((0), dtype=np.float32)
n_pre = 200
for i in tqdm(range(n_pre)):
    arima_3_1_0 = ARIMA(ts, order=(3, 1, 0)).fit(dist=False)
    predict = arima_3_1_0.forecast()[0]
    predictions = np.hstack([predictions, predict])
    ts = np.hstack([ts, predict])
predictions
```

```python
nans = np.zeros(len(train))
nans[:] = np.nan
orgs = pd.concat([train['AAL_Close'], test[:n_pre]['AAL_Close']])
orgs = pd.DataFrame({
    'Date': orgs.index,
    'Original': orgs.as_matrix(),
    #'Prediction': np.hstack([nans, predictions])
})
orgs = orgs.set_index('Date')
orgs.plot(color=['blue', 'red'])
plt.show()
```
